```python
# Import libraries
# SQLAlchemy
from sqlalchemy import create_engine
from sqlalchemy import inspect
from config import db_password

# Pandas
import pandas as pd
from collections import Counter

import matplotlib.pyplot as plt
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.metrics import confusion_matrix
from sklearn.metrics import balanced_accuracy_score
from imblearn.metrics import classification_report_imbalanced
from sklearn.metrics import classification_report
from imblearn.ensemble import EasyEnsembleClassifier

from sklearn.svm import SVC
```

```python
file = 'clean_ml_data.csv'
clean_df = pd.read_csv(file)
```

```python
clean_df
```

```python
y= clean_df['intubation']
y1 = clean_df['ICU']
y2 = clean_df['death']
X = clean_df.drop(columns = ['ICU','intubation','death'])
```

### Use Easy Ensemble Classifier - on 'ICU', 'INTUBATION', 'DEATH'

```python
X0_train,X0_test,y0_train,y0_test = train_test_split(X,y,random_state =42, stratify=y)
```

```python
X1_train,X1_test,y1_train,y1_test = train_test_split(X,y1,random_state =42, stratify=y)
```

```python
X2_train,X2_test,y2_train,y2_test = train_test_split(X,y2,random_state =42, stratify=y)
```

```python
# Creating StandardScaler instance
scaler = StandardScaler()
# Fitting standard scaler
X0_scaler = scaler.fit(X0_train)

# Scaling data
X0_train_scaled = X0_scaler.transform(X0_train)
X0_test_scaled = X0_scaler.transform(X0_test)
```

```python
# Creating StandardScaler instance
scaler = StandardScaler()
# Fitting standard scaler
X1_scaler = scaler.fit(X1_train)
# Scaling data
X1_train_scaled = X1_scaler.transform(X1_train)
X1_test_scaled = X1_scaler.transform(X1_test)
```

```python
# Creating StandardScaler instance
scaler = StandardScaler()
# Fitting standard scaler
X2_scaler = scaler.fit(X2_train)
# Scaling data
X2_train_scaled = X2_scaler.transform(X2_train)
X2_test_scaled = X2_scaler.transform(X2_test)
```

```python
# Train on the EasyEnsembleClassifier
from imblearn.ensemble import EasyEnsembleClassifier
eec = EasyEnsembleClassifier(random_state=1,n_estimators = 100)
eec.fit(X0_train_scaled,y0_train)


# Calculated the balanced accuracy score
y0_pred_eec = eec.predict(X0_test_scaled)

print(f"The target 0 's balanced_accuracy_score is: {balanced_accuracy_score(y0_test, y0_pred_eec)}")
```

```python
# Train on the EasyEnsembleClassifier
from imblearn.ensemble import EasyEnsembleClassifier
eec = EasyEnsembleClassifier(random_state=1,n_estimators = 100)

eec.fit(X1_train_scaled,y1_train)

# Calculated the balanced accuracy score

y1_pred_eec = eec.predict(X1_test_scaled)

print(f"The target 1 's balanced_accuracy_score is: {balanced_accuracy_score(y1_test, y1_pred_eec)}")
```

```python
# Train on the EasyEnsembleClassifier
from imblearn.ensemble import EasyEnsembleClassifier
eec = EasyEnsembleClassifier(random_state=1,n_estimators = 100)

eec.fit(X2_train_scaled,y2_train)

# Calculated the balanced accuracy score

y2_pred_eec = eec.predict(X2_test_scaled)

print(f"The target 2 's balanced_accuracy_score is: {balanced_accuracy_score(y2_test, y2_pred_eec)}")
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y0_test, y0_pred_eec))
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y1_test, y1_pred_eec))
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y2_test, y2_pred_eec))
```
